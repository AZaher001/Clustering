```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(rc={'figure.figsize': [10, 11]}, font_scale=1.3)
```

```python
df = pd.read_csv('Country-data.csv')
df
```

```python
df['country'].unique()
```

```python
x = df.drop('country',axis = 1)
x
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_scaler = scaler.fit_transform(x)
```

# Cluster _ KMean

```python
from sklearn.cluster import KMeans
model= KMeans(n_clusters=3)
cluster = model.fit_predict(df_scaler)
cluster
```

```python
# choose the best number of clusters (K Value) using Elbow Method
scores = []
for i in range(1, 70):
    model = KMeans(n_clusters=i)
    model.fit(df_scaler)
    scores.append(model.inertia_)
plt.plot(range(1, 70), scores)
plt.title("Elbow Method")
plt.xlabel("num of clusters")
plt.ylabel("Score")
```

```python
model = KMeans(n_clusters=40)
cluster_update = model.fit_predict(df_scaler)
cluster_update
```

```python
df['KM_cluster'] = cluster_update
df
```

# Cluster _ Hierarchical

```python
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
```

```python
dendrogram = sch.dendrogram(sch.linkage(x, method='ward'))
```

```python
model = AgglomerativeClustering(n_clusters=13)
cluster_df = model.fit_predict(df_scaler)
cluster_df
```

```python
df['sch_Cluster'] = cluster_df
df
```

```python
y_data = df.sort_values(by='country')
for i in range(40) :
    print(f"--- KM_cluster : {i} ")
    print(y_data[y_data['KM_cluster'] == i]['country'])
    print('---------------------------------------')
```
